```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random

def Random_image_dispaly(target_dir,target_class):
  target_path = target_dir + target_class
  random_image = random.sample(os.listdir(target_path),1)
  print(random_image)
  img = mpimg.imread(target_path + "/" + random_image[0])
  # print("Image Shape : ",img.shape)
  plt.imshow(img)
  plt.title(target_class)
  plt.axis(False)
  return img
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

tf.random.set_seed(42)

train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_dir = "/content/pizza_steak/train"
test_dir = "/content/pizza_steak/test"

traindata = train_datagen.flow_from_directory(directory=train_dir,batch_size=32,class_mode="binary",target_size=(224,224),seed=42)
test_data = train_datagen.flow_from_directory(directory=test_dir,batch_size=32,class_mode="binary",target_size=(224,224),seed=42)
```

```python
image_model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters=10,kernel_size=(3,3),activation="relu",input_shape=(224,224,3)),
    tf.keras.layers.Conv2D(10,3,activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=2,padding="valid"),
    tf.keras.layers.Conv2D(10,3,activation="relu"),
    tf.keras.layers.Conv2D(10,3,activation="relu"),
    tf.keras.layers.MaxPool2D(2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1,activation="sigmoid")
])
image_model.compile(loss=tf.keras.losses.binary_crossentropy,optimizer=tf.keras.optimizers.Adam(),metrics=["accuracy"])
history = image_model.fit(traindata,epochs=4,steps_per_epoch=len(traindata),validation_data=test_data,validation_steps=len(test_data))
```

```python
image_model_4 = tf.keras.Sequential([
    tf.keras.layers.Conv2D(10,3,1,activation="relu",input_shape=(224,224,3)),
    tf.keras.layers.Conv2D(10,3,1,activation="relu"),
    tf.keras.layers.AvgPool2D(2),
    tf.keras.layers.Conv2D(10,3,1,activation="relu"),
    tf.keras.layers.AvgPool2D(2),
    tf.keras.layers.Conv2D(10,3,1,activation="relu"),
    tf.keras.layers.AvgPool2D(2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

image_model_4.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.Adam(),metrics=["accuracy"])
history_4 = image_model_4.fit(train_data,epochs=4,steps_per_epoch=len(train_data),validation_data=test_data,validation_steps=len(test_data))
```

```python
image_model_5 = tf.keras.Sequential([
    tf.keras.layers.Conv2D(10,3,1,activation="relu",input_shape=(224,224,3)),
    tf.keras.layers.Conv2D(10,3,1,activation="relu"),
    tf.keras.layers.AvgPool2D(2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

image_model_5.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.Adam(),metrics=["accuracy"])
history_5 = image_model_5.fit(train_data,epochs=4,steps_per_epoch=len(train_data),validation_data=test_data,validation_steps=len(test_data))
```

```python
h4 = pd.DataFrame(history_4.history)
h5 = pd.DataFrame(history_5.history)
h4.plot(title="Image Model 4")
h5.plot(title="Image Model 5")
```
